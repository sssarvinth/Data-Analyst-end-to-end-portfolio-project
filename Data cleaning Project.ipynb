```python
## Let us create a dummy variable for the reason for absence column, this will create a table which as 29 columns with each column mapped for one reason


## we can achieve this using the get dummy function


reason_columns=pd.get_dummies(df['Reason for Absence'])
reason_columns
```

```python
## to validate on the number of reason shared by a person for leave, let us create a new column and sum up the rows.

##if value is 1 then, only 1 reason is given by the employee
##if value is 0, then no reason is given by the employee
##if value is > 1, then more than 1 reason is given by the employee.


## let us create a check column to validate tis


reason_columns['check']=reason_columns.sum(axis=1)
reason_columns
```

```python
## observation, we see only 1 to occur in most cases, we can confirm that only 1 reason is shared by the employees for  information

## to prove this using python we can use the unique function in check column

reason_columns['check'].unique()
```

```python
##based on the observations we can infer that employees are giving only 1 reason for opting leaves. We can opt to drop 
```

```python
## as we have infered from the logic of the reasons_columns, we can now drop the reaons_columns

reason_columns=reason_columns.drop(['check'],axis=1)
reason_columns
```

```python
## Drop the reason 0

reason_columns=pd.get_dummies(df['Reason for Absence'],drop_first=True)
reason_columns
```

```python
## Grouping the reasons for absence

df.columns.values
```

```python
df=df.drop(['Reason for Absence'],axis=1)
```

```python
## Our goal is to group the reasons and simplify the dataframe

## reasons 1-14 are all related to diseases
##reasons 15-17 are all related to pregancy and giving birth
## reasons 18-20 are all related to poisions
## reasons 22-28 are all related to casual reasons such as medical appointment , consultation etc


reason_type_1 = reason_columns.loc[:,1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:,18:20].max(axis=1)
reason_type_4 = reason_columns.loc[:,22:].max(axis=1)


```

```python
reason_type_1
```

```python
##Concatenate column values

df=pd.concat([df, reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
df
```

```python
df_reason_mod['Date']=pd.to_datetime(df_reason_mod['Date'],format='%d/%m/%Y')
df_reason_mod['Date']
```

```python
## create a list for the month number and map it to the main dataframe

list_months=[]
list_months
for i in range(700):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
## Extract days of the week

def date_to_days_of_the_week(value):
    return value.weekday()

```

```python
## create a day of the week column and apply the above function

df_reason_mod['Day of the week']=df_reason_mod['Date'].apply(date_to_days_of_the_week)
```
